```python
df1=df.drop('indicator_name',axis=1)
```

```python
df1
```

```python
df2=df1.drop('age_group',axis=1)
```

```python
df2
```

```python
y=df2.iloc[:,0]
```

```python
x=df2.iloc[:,1:]
```

```python
from sklearn.preprocessing import LabelEncoder 
```

```python
label_encoder= LabelEncoder()
```

```python
x["sex"]=label_encoder.fit_transform(x["sex"])
```

```python
x["age_categories"]=label_encoder.fit_transform(x["age_categories"])
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/4,random_state=0)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
print(y_test)
```

```python
print(y_train)
```

```python
import sklearn.tree as tree
clf=tree.DecisionTreeClassifier(max_depth=3,random_state=200)
clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
clf.score(x_test,y_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
plt.figure(figsize=(7, 6)) 
cm=confusion_matrix(y_test, y_pred)
import seaborn as sns
sns.heatmap(cm,annot=True, cmap='Blues') 
plt.title('Confusion Matrix')
plt.xlabel('Predicted Activity')
plt.ylabel('Actual Activity')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

RFC = RandomForestClassifier(n_estimators =40, random_state = 0)
RFC.fit(x_train,y_train)

x_pred = RFC.predict(x_test)
```

```python
# Define the logistic function
def logistic_function(x):
    return 1 / (1 + np.exp(-x))

# Generate x values
x_values = np.linspace(-6, 6, 400)

# Calculate corresponding y values using the logistic function
y_values = logistic_function(x_values)

# Set the style to white background with dark grid
sns.set(style="white")

# Create a figure and axis with a larger size
plt.figure(figsize=(10, 6))

# Plot the logistic function with colorful line and shaded area
plt.plot(x_values, y_values, label='Logistic Function', color='purple', linewidth=2)

# Add labels and title
plt.title('Logistic Function', fontsize=16)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)

# Customize tick labels
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Add a legend
plt.legend(fontsize=12)

# Add a shaded region to highlight the area under the curve
plt.fill_between(x_values, y_values, color='lightblue', alpha=0.4)

# Set background color to white
ax = plt.gca()
ax.set_facecolor('white')

# Remove spines
sns.despine()

# Display the plot
plt.tight_layout()
plt.show()
```
